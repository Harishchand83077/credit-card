```python
import numpy as np
import pandas as pd
```

```python
file_path = "/content/DOGE-USD.csv"
df = pd.read_csv(file_path)
```

```python
df.head(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.corr
```

```python
# to set date time as index in series data
df['Date'] = pd.to_datetime(df['Date'],infer_datetime_format=True)
df.set_index('Date', inplace=True)

df.isnull().any()
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.describe()
```

```python
plt.figure(figsize=(20,7))
x=df.groupby('Date')['Close'].mean()
x.plot(linewidth=2.5, color='b')
plt.xlabel("Date")
plt.ylabel("volume")
plt.title("date vs volume")
plt.show()
```

```python
df['gap']=(df['High']-df['Low'])*df['Volume']
df['y']=df['High']/df['Volume']
df['z']=df['Low']/df['Volume']
df['a']=df['High']/df['Low']
df['b']=(df['High']/df['Low'])*df['Volume']
abs(df.corr()['Close'].sort_values(ascending=False))
```

```python
df = df[["Close", "Volume", "gap", "a", "b"]]
df.head()
```

```python
df2 = df.tail(30)
train = df2[:11]
test = df2[-19:]

print(train.shape, test.shape)
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
model = SARIMAX(endog=train["Close"], exog=train.drop(
    "Close", axis=1), order=(2, 1, 1))
results = model.fit()
print(results.summary())
```

```python
start=11
end=29
forecast = results.predict(start=start, end=end, exog=test.drop("Close", axis=1))
forecast
```

```python
test["Close"].plot(legend=True, figsize=(12, 6))
forecast.plot(label='TimeSeries', legend=True)
```
